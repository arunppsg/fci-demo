```python
def make_to_frame(to_frame):
    to_frame = pd.DataFrame(to_frame)
    to_frame = to_frame[to_frame['State/UT'] != 'TOTAL']
    to_frame = to_frame[to_frame['State/UT'] != 'GRAND TOTAL']
    return to_frame
```
